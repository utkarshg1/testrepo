# Lasso Regularization (L1 Regularization)
Loss function penalty is applied on Absolute value of coefficents

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Step 1: Read the dataset

```python
import pandas as pd
df = pd.read_csv('Cars93.csv', na_values=['', 'NA'], keep_default_na=False)
df.head()
```

### Step 2: Perform basic data quality checks

```python
df.info()
```

```python
m = df.isna().sum()
m
```

```python
m[m>0]
```

```python
df.duplicated().sum()
```

### Step 3 : Seperate X and Y(Weight)

```python
X = df.drop(columns=['id', 'Weight'])
Y = df[['Weight']]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4: Perform preprocessing on X

```python
X.columns
```

```python
X.dtypes
```

```python
# seperate categorical and continous features
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = Pipeline(steps=[('imputer', SimpleImputer(strategy='mean')),
                           ('scaler', StandardScaler())])
```

```python
cat_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='most_frequent')),
                           ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre = ColumnTransformer([('num', num_pipe, con),
                         ('cat', cat_pipe, cat)]).set_output(transform='pandas')
```

```python
pre
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

### Apply train test split
20% - test

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=60)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

### Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

### Evaluate above model

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Build a Lasso model(L1 regularization)

```python
from sklearn.linear_model import Lasso
model2 = Lasso(alpha=5)
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain, ytrain)
```

```python
model2.score(xtest, ytest)
```

### Hyperparameter tuning with GridSearchCV

```python
import numpy as np 
params = {'alpha':np.arange(start=0.1, stop=100, step=0.1)}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
ls = Lasso()
gscv = GridSearchCV(estimator=ls, param_grid=params, cv=5, scoring='neg_mean_squared_error')
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_
```

```python
gscv.best_score_
```

```python
best_lasso = gscv.best_estimator_
best_lasso
```

### Evaluate best lasso model

```python
best_lasso.score(xtrain, ytrain)
```

```python
best_lasso.score(xtest, ytest)
```

### Get the cross validated r2 score on train

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(best_lasso, xtrain, ytrain, cv=5, scoring='r2')
scores
```

```python
scores.mean()
```

```python
scores2 = cross_val_score(best_lasso, xtrain, ytrain, cv=5, scoring='neg_mean_absolute_error')
scores2
```

```python
scores2.mean()
```

### Above model has good r2 score in test and also in cross validation

```python
ypred_train = best_lasso.predict(xtrain)
ypred_test = best_lasso.predict(xtest)
```
